```python
import pandas as pd 

file_path = "C:/Users/acer/OneDrive/Desktop/marketing_campaign.csv"

df = pd.read_csv(file_path, sep = '\t')

df
```

```python
df.isnull().sum()
```

```python
df.fillna(method = 'ffill' , inplace = True)

df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df.rename(columns={
    'kidhome': 'kid_home',
    'Teenhome': 'Teen_home',
    'Dt_Customer': 'Date',
    'MntWines': 'Mnt_Wines',
    'MntFruits': 'Mnt_Fruits',
    'MntMeatProducts': 'Mnt_Meat_Products',
    'MntFishProducts': 'Mnt_Fish_Products',  # removed extra space before Mnt
    'MntGoldProds': 'Gold_Products',
    'NumDealPurchases': 'Deal_Purchases',
    'NumWebPurchases': 'Web_Purchases',
    'NumCatalogPurchases': 'Catalog_Purchases',
    'NumStorePurchases': 'Store_Purchases',
    'NumWebVisitsMonth': 'Web_Visit_Month',
    'AcceptedCmp3': 'Cmp3',
    'AcceptedCmp4': 'Cmp4',
    'AcceptedCmp1': 'Cmp1',
    'AcceptedCmp2': 'Cmp2',
    'AcceptedCmp5' : 'Cmp5'
}, inplace=True)


df.columns
                                  
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

df['Education'] = df['Education'].astype('category')
df['Marital_Status'] = df['Marital_Status'].astype('category')


binary_cols = ['Cmp3', 'Cmp4', 'Cmp5', 'Cmp1', 'Cmp2', 'Complain', 'Response']

df[binary_cols] = df[binary_cols].astype(bool)

df.info()
```

```python
df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y', errors='coerce')
df['Date'] = df['Date'].fillna(df['Date'].min())

df
```
